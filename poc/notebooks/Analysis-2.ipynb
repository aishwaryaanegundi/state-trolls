# Analysis of hits of relaxed threshold

```python
months = ['2016-03', '2016-04', '2016-05', '2016-06',
         '2016-07', '2016-08', '2016-09', '2016-10', '2016-11', '2016-12']
#          '2017-01', '2017-02', '2017-03','2017-04', '2017-05', '2017-06']
```

```python
import datetime
top_authors_complete = []
for m in months:
    hits_relaxed = pd.read_csv('/INET/state-trolls/work/state-trolls/poc/stsb-roberta-large/results/relaxed_threshold/'+m+'.csv')
    hits_reduced = hits_relaxed.drop_duplicates(subset=['tweet_id','post_id'], keep='last')
    print(hits_relaxed.shape, hits_reduced.shape)
    print(hits_relaxed.columns)
    hits_processed = get_relevant_hits(m, hits_reduced)
    print('############',hits_processed.shape)
    hits_processed['abs_td'] = hits_processed['td'].abs()
    hits_per_author = hits_processed.groupby(['post_author'])["post_id"].count().reset_index(name="count")
    hits_count_5 = hits_per_author[hits_per_author['count']>5]
    avg_tds = []
    for a in hits_count_5['post_author'].tolist():
        rel_hits = hits_processed[hits_processed['post_author']==a]
        avg_td = np.mean(rel_hits['abs_td'].tolist())
        avg_tds.append(avg_td)
    hits_count_5['avg_td'] = avg_tds    
    hits_count_5 = hits_count_5.sort_values('avg_td', ascending=False)
    top_1000 = hits_count_5.head(10)
    top_1000 = top_1000.sort_values('count', ascending=False)
    top_10_authors = top_1000['post_author'].tolist()
    for a in top_10_authors:
        top_authors_complete.append(a)
    
```

```python
def get_relevant_hits(month, hits):
    combined_hits = pd.DataFrame()
    reddit_data = pd.read_csv('/INET/state-trolls/work/state-trolls/reddit_dataset/comments/posts/posts-'+ month +'.csv')  
    fdata = data[data['tweetid'].isin(hits['tweet_id'].to_list())]
    fposts = reddit_data[reddit_data['id'].isin(hits['post_id'].to_list())]
    hits_processed = []
    for index, row in hits.iterrows():
        post = fposts[fposts['id'] == row['post_id']]
        if not (post.empty):
            try:
                tweet = fdata[fdata['tweetid'] == row['tweet_id']]
                element = fdata[fdata['tweetid'] == row['tweet_id']]
                tweet_time = (element['tweet_time'].values)[0]
                post_time = (post['created_utc'].values)[0]
                post_times = post_time.split()
                post_time = post_times[0]
                post_time = (datetime.datetime.fromtimestamp(int(post_time)))
                
                
#                 print('adding')
                hits_processed.append({'tweet_id': row['tweet_id'], 'post_id': row['post_id'], 
                             'post_author': (post['author'].values)[0],
                             'post_time': post_time, 'tweet_time': tweet_time})
            except IndexError:
                print(post['created_utc'])
            except ValueError:
                print('Its a value error')
                print(post['created_utc'])
                print(post_time)
            except TypeError:
                print(post['created_utc'])
                post_times = post_time.split()
#                 print('post times: ', post_times)
                post_time = post_times[0]
            except AttributeError:
                print(post['created_utc'])
                hits_processed.append({'tweet_id': row['tweet_id'], 'post_id': row['post_id'], 
                             'post_author': (post['author'].values)[0],
                             'post_time': None, 'tweet_time': tweet_time})
            except Exception as ex:
                print('createdutc: ', post['created_utc'])
                print(ex)
                post_times = post_time.split()
                print('post times: ', post_times)
                post_time = post_times[1]

#     print('list length: ', len(hits_processed))
    hits_df = pd.DataFrame(hits_processed)
    hits_df = hits_df[hits_df['post_time'].notnull()]
    if not (hits_df.empty):
        t = pd.to_datetime(hits_df.tweet_time, infer_datetime_format = True)
        del hits_df['tweet_time']
        hits_df['tweet_time'] = t
        hits_df['td'] = (hits_df.post_time - hits_df.tweet_time).astype('timedelta64[s]')
#     print('before returning........',hits_df.shape)
    return hits_df
```
