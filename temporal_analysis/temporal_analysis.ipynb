```python
import numpy as np 
import pandas as pd
import csv
import os
import matplotlib.pyplot as plt

# displays all columns and rows when asked to print
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
path = "../datasets/russia_052020_tweets_csv_hashed_2.csv"
```

```python
data = pd.read_csv(path, sep=',')
```

```python
data['tweet_time'].dtype
data['tweet_time'] = pd.to_datetime(data['tweet_time'])
```

```python
# extracts just the english tweets by using the l
is_english_tweet = data['tweet_language'] == 'en'
english_data = data[is_english_tweet]
print(english_data.shape)

# english_data = data
```

```python
data = data.rename(columns={'date': 'time'})
data['date'] = pd.to_datetime(data['tweet_time'], format='%H:%M').dt.date
```

```python
tempo_data = data.groupby(['date']).size()
tempo_data=tempo_data.to_frame().reset_index()
tempo_data.columns = ['date', 'count']
```

```python
tempo_data.plot( kind='line', x='date', y='count',title='Number of tweets')
axes = plt.gca()
plt.show()
```
